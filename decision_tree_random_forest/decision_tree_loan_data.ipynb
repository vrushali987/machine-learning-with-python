```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('loan_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
purpose1=pd.get_dummies(df['purpose'],drop_first=True)
```

```python
purpose1
```

```python
df=pd.concat([df,purpose1],axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop('purpose',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(),cmap='coolwarm')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
scaled_array=ss.fit_transform(df.drop('not.fully.paid',axis=1))
```

```python
scaled_array.shape
```

```python
l=list(df.columns)
```

```python
l
```

```python
l.remove('not.fully.paid')
```

```python
X=pd.DataFrame(data=scaled_array,columns=l)
```

```python
X.head()
```

```python
y=df['not.fully.paid']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print (confusion_matrix(y_test,y_pred))
```
